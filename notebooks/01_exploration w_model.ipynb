```python
import torch
import torchvision
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
```

```python
from torchvision import transforms

train_tfms = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

val_tfms = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])
```

## Datasets

```python
dataset = datasets.ImageFolder(
    root="/content/goatvision_dataset",
    transform=train_tfms
)
```

## Split

```python
from torch.utils.data import random_split

train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size

train_dataset, val_dataset = random_split(dataset,[train_size,val_size])

val_dataset.dataset.transform = val_tfms
```

## DataLoader

```python
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=16)
```

## Model

```python
import torchvision.models as models
import torch.nn as nn

model = models.resnet50(pretrained=True)

num_features = model.fc.in_features
model.fc = nn.Linear(num_features, 5)
```

## Optimizer

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.0001)
```

## Training

```python
epochs = 5

for epoch in range(epochs):

    model.train()
    running_loss = 0

    for images, labels in train_loader:

        outputs = model(images)
        loss = criterion(outputs, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {running_loss}")
```

```python
correct = 0
total = 0

model.eval()

with torch.no_grad():
    for images, labels in val_loader:

        outputs = model(images)
        _, predicted = torch.max(outputs, 1)

        total += labels.size(0)
        correct += (predicted == labels).sum().item()

accuracy = 100 * correct / total
print("Validation Accuracy:", accuracy)
```

## High-Confidence Mismatch Finder

```python
import torch
import numpy as np
from torch.utils.data import DataLoader
import os

# Make sure model uses GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
model.eval()

# Use a non-shuffled loader so indices line up
audit_loader = DataLoader(dataset, batch_size=32, shuffle=False)

suspects = []  # (confidence, true_label, pred_label, filepath)

with torch.no_grad():
    idx = 0
    for images, labels in audit_loader:
        images = images.to(device)
        labels = labels.to(device)

        outputs = model(images)
        probs = torch.softmax(outputs, dim=1)
        confs, preds = torch.max(probs, dim=1)

        confs = confs.cpu().numpy()
        preds = preds.cpu().numpy()
        labels_np = labels.cpu().numpy()

        # dataset.samples holds (filepath, class_index) in the same order as dataset indexing
        for b in range(len(labels_np)):
            true_i = labels_np[b]
            pred_i = preds[b]
            conf = float(confs[b])

            if pred_i != true_i and conf >= 0.85:  # adjust threshold if needed
                filepath, _ = dataset.samples[idx + b]
                suspects.append((conf, dataset.classes[true_i], dataset.classes[pred_i], filepath))

        idx += len(labels_np)

# Sort most confident mistakes first
suspects.sort(reverse=True, key=lambda x: x[0])

print(f"Found {len(suspects)} high-confidence mismatches (conf >= 0.85).")
print("Top 15:")
for item in suspects[:15]:
    conf, true_lbl, pred_lbl, fp = item
    print(f"{conf:.2f} | true={true_lbl:12s} pred={pred_lbl:12s} | {os.path.basename(fp)}")
```

```python
from PIL import Image
import matplotlib.pyplot as plt

def show_suspects(suspects, n=12):
    n = min(n, len(suspects))
    plt.figure(figsize=(14, 8))
    for i in range(n):
        conf, true_lbl, pred_lbl, fp = suspects[i]
        img = Image.open(fp).convert("RGB")
        plt.subplot(3, 4, i+1)
        plt.imshow(img)
        plt.title(f"{conf:.2f}\ntrue:{true_lbl}\npred:{pred_lbl}", fontsize=9)
        plt.axis("off")
    plt.tight_layout()
    plt.show()

show_suspects(suspects, n=12)
```

## Confusion Matrix

```python
import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

model.eval()
all_preds = []
all_labels = []

with torch.no_grad():
    for images, labels in val_loader:
        outputs = model(images)
        preds = outputs.argmax(dim=1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

cm = confusion_matrix(all_labels, all_preds)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=dataset.classes)

plt.figure(figsize=(7,7))
disp.plot(values_format='d')
plt.xticks(rotation=45, ha="right")
plt.title("G.O.A.T Vision Confusion Matrix")
plt.show()
```

```python
import torch
import torch.nn as nn
import torchvision.models as models

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

# Freeze all parameters
for p in model.parameters():
    p.requires_grad = False

# Unfreeze only layer4 + final FC
for p in model.layer4.parameters():
    p.requires_grad = True
for p in model.fc.parameters():
    p.requires_grad = True
```

```python
criterion = nn.CrossEntropyLoss()

optimizer = torch.optim.Adam([
    {"params": model.layer4.parameters(), "lr": 1e-5},
    {"params": model.fc.parameters(),     "lr": 1e-4},
])
```

```python
epochs = 5
for epoch in range(epochs):
    model.train()
    running = 0.0
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running += loss.item() * images.size(0)

    print(f"FT Epoch {epoch+1}/{epochs} | train_loss={running/len(train_loader.dataset):.4f}")
```

```python
import torch
import torch.nn as nn

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.eval()

criterion = nn.CrossEntropyLoss()

val_loss = 0.0
correct = 0
total = 0

with torch.no_grad():
    for images, labels in val_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)

        loss = criterion(outputs, labels)
        val_loss += loss.item() * images.size(0)

        preds = outputs.argmax(dim=1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

print(f"Val loss: {val_loss/total:.4f}")
print(f"Val accuracy: {100*correct/total:.2f}%")
```

```python
from pytorch_grad_cam import GradCAM
from pytorch_grad_cam.utils.model_targets import ClassifierOutputTarget
from pytorch_grad_cam.utils.image import show_cam_on_image
import numpy as np
import cv2

# choose last convolution layer
target_layers = [model.layer4[-1]]

cam = GradCAM(model=model, target_layers=target_layers)

images, labels = next(iter(val_loader))

input_tensor = images[0].unsqueeze(0)

targets = [ClassifierOutputTarget(labels[0].item())]

grayscale_cam = cam(input_tensor=input_tensor, targets=targets)
grayscale_cam = grayscale_cam[0, :]

img = images[0].permute(1,2,0).cpu().numpy()

# undo normalization for display
img = (img - img.min()) / (img.max() - img.min())

visualization = show_cam_on_image(img, grayscale_cam, use_rgb=True)

plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("G.O.A.T Vision Attention")
plt.imshow(visualization)
plt.axis("off")

plt.show()
```
